```python
print("Enter the number of queens")
N = int(input())
board = [[0]*N for _ in range(N)]
def attack(i, j):
       for k in range(0, N):
        if board[i][k] == 1 or board[k][j] == 1:
            return True
    # Checking diagonally
       for k in range(0, N):
        for l in range(0, N):
            if (k + l == i + j) or (k - l == i - j):
                if board[k][l] == 1:
                    return True
       return False

def N_queens(n):
    if n == 0:
        return True
    for i in range(0, N):
        for j in range(0, N):
            if (not attack(i, j)) and (board[i][j] != 1):
                board[i][j] = 1
                if N_queens(n - 1):
                    return True
                board[i][j] = 0
    return False

N_queens(N)

for i in board:
    print(i)
```

```python
def print_solution(board):
    for row in board:
        print("  ".join(row))
    print("\n")

def is_safe(board, row, col, n):
    # Check for queens in the same column
    for i in range(row):
        if board[i][col] == "1":
            return False
    # Check for queens in the upper left diagonal
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == "1":
            return False
    # Check for queens in the upper right diagonal
    for i, j in zip(range(row, -1, -1), range(col, n)):
        if board[i][j] == "1":
            return False
    return True

def solve_nqueens_util(board, row, n, solution_count):
    if solution_count[0] == 10:  # Stop when 10 solutions are found
        return
    if row == n:
        print_solution(board)
        solution_count[0] += 1
        return
    for col in range(n):
        if is_safe(board, row, col, n):
            board[row][col] = "1"
            solve_nqueens_util(board, row + 1, n, solution_count)
            board[row][col] = "."  # Backtrack

def solve_nqueens(n):
    board = [["." for _ in range(n)] for _ in range(n)]
    solution_count = [0]
    solve_nqueens_util(board, 0, n, solution_count)
    if solution_count[0] == 0:
        print("No solution exists.")
    elif solution_count[0] < 10:
        print(f"Found {solution_count[0]} solutions.")
    else:
        print("Found 10 solutions.")

n = int(input("Enter the size of the chessboard (N): "))
solve_nqueens(n)
```
