```python
import numpy as np
import pandas as pd

def learn(concepts, target):
    specific_h = concepts[0].copy()
    general_h = [["?" for i in range(len(specific_h))] for i in range(len(specific_h))]
    print("\nInitial Specific Hypothesis: ", specific_h)
    print("\nInitial Generic Hypothesis: ",general_h)  

    for i, h in enumerate(concepts):
        print("\nInstance", i+1 , "is ", h)
        if target[i] == "yes":
            print(f"Instance {i+1} is Positive ")
            for x in range(len(specific_h)):
                if h[x]!= specific_h[x]:                    
                    specific_h[x] ='?'                    
                    general_h[x][x] ='?'
                   
        if target[i] == "no":            
            print(f"Instance {i+1} is Negative ")
            for x in range(len(specific_h)):
                if h[x]!= specific_h[x]:                    
                    general_h[x][x] = specific_h[x]                
                else:                    
                    general_h[x][x] = '?'        
       
        print("Specific Hypothesis after Instance", i+1,":\n",specific_h)        
        print("Generic Hypothesis after Instance", i+1,":\n", general_h)
        print("\n")

    indices = [i for i, val in enumerate(general_h) if val == ['?', '?', '?', '?', '?', '?']]    
    for i in indices:  
        general_h.remove(['?', '?', '?', '?', '?', '?'])
    return specific_h, general_h
data = pd.read_csv('data.csv')
df = pd.DataFrame(data)
print("\nData:\n",df)
concepts = np.array(data.iloc[:,0:-1])
target = np.array(data.iloc[:,-1])

s_final, g_final = learn(concepts, target)

print("Final Specific Hypothesis: ", s_final, sep="\n")
print("Final General Hypothesis: ", g_final, sep="\n")

```
