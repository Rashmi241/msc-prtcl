```python
def create_node(key):
    return {"key": key, "left": None, "right": None}

def insert(root, key):
    if root is None:
        return create_node(key)
    if key == root["key"] :
        print("Element already exists!")
    if key < root["key"]:
        root["left"] = insert(root["left"], key)
    else:
        root["right"] = insert(root["right"], key)
    return root

def min_value_node(root):
    if root is None:
        return None
    while root["left"] is not None:
        root = root["left"]
    return root

def delete(root, key):
    if root is None:
        print(f"Key {key} is not present in the BST.")
        return root

    if key < root["key"]:
        root["left"] = delete(root["left"], key)
    elif key > root["key"]:
        root["right"] = delete(root["right"], key)
    else:
        # Node with only one child or no child
        if root["left"] is None:
            return root["right"]
        elif root["right"] is None:
            return root["left"]

        # Node with two children
        min_right = min_value_node(root["right"])  # Find the minimum value node in the right subtree
        root["key"] = min_right["key"]  # Replace the key of the current node with the minimum value from the right subtree
        root["right"] = delete(root["right"], min_right["key"])  # Delete the minimum value node from the right subtree

    return root



def search(root, key):
    if root is None or root["key"] == key:
        return root
    if key < root["key"]:
        return search(root["left"], key)
    return search(root["right"], key)

def inorderTraversal(root):
    if root:
        inorderTraversal(root["left"])
        print(root["key"], end=" ")  # Print the node's key
        inorderTraversal(root["right"])

def preorderTraversal(root):
    if root:
        print(root["key"], end=" ") 
        preorderTraversal(root["left"])
        preorderTraversal(root["right"])

def postorderTraversal(root):
    if root: 
        postorderTraversal(root["left"])
        postorderTraversal(root["right"])
        print(root["key"], end=" ")

def main():
    root = None
    print("BST operations :")
    print("1. Insert")
    print("2. Delete")
    print("3. Search")
    print("4. Inorder Traversal")
    print("5. Preorder Traversal")
    print("6. Postorder Traversal")
    print("7. Exit")
    print("\n")

    while True :
      ch=int(input("\nEnter your choice :"))

      if ch == 1:
           key = int(input("Enter an element to insert: "))
           root = insert(root, key)
      elif ch ==2:
           key = int(input("Enter the element to be deleted: "))
           root = delete(root, key)
      elif ch ==3:
           search_key = int(input("Enter a key to search for: "))
           result = search(root, search_key)
           if result :
               print(f"Key {search_key} found in the BST.")
           else :
               print(f"Key {search_key} not found in the BST.")
      elif ch == 4:
           print("Inorder traversal is:", end=" ")
           inorderTraversal(root)
      elif ch == 5:
           print("preorder traversal is:", end=" ")
           preorderTraversal(root)
      elif ch == 6:
           print("postorder traversal is:", end=" ")
           postorderTraversal(root)
    
      elif ch==7 :
           print("*** Exit ***")
           break
      else :
          print("Invalid Choice.Enter your choice between 1 and6 only ") 
     
               
if __name__ == "__main__":
    main()
```
